```python
%pip install selenium chromedriver-autoinstaller

from selenium import webdriver
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller
import time

chromedriver_autoinstaller.install()
```

```python
def get_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless=new')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    driver.set_page_load_timeout(30)
    return driver

def test_onet():
    d = get_driver()
    d.get('https://www.onet.pl')
    assert 'Onet' in d.title
    assert d.find_element(By.TAG_NAME, 'header')
    assert d.find_element(By.TAG_NAME, 'nav').is_displayed()
    d.quit()

def test_example():
    d = get_driver()
    d.get('https://example.com')
    assert d.title == 'Example Domain'
    assert 'Example Domain' in d.find_element(By.TAG_NAME, 'h1').text
    assert 'iana.org' in d.find_element(By.TAG_NAME, 'a').get_attribute('href')
    d.quit()

def test_wikipedia():
    d = get_driver()
    d.get('https://pl.wikipedia.org')
    assert d.find_element(By.NAME, 'search')
    assert d.find_element(By.ID, 'n-randompage').is_displayed()
    links = d.find_elements(By.CSS_SELECTOR, 'a[href*="/wiki/"]')
    assert len(links) > 0
    d.quit()

def test_github():
    d = get_driver()
    d.get('https://github.com')
    assert 'GitHub' in d.title
    assert d.find_element(By.CSS_SELECTOR, 'a[href*="signup"]')
    assert d.find_element(By.TAG_NAME, 'header')
    d.quit()

def test_google():
    d = get_driver()
    d.get('https://www.google.com')
    assert d.find_element(By.CSS_SELECTOR, 'img[alt="Google"]')
    assert d.find_element(By.NAME, 'q')
    assert d.find_element(By.TAG_NAME, 'form')
    d.quit()

def test_youtube():
    d = get_driver()
    d.get('https://www.youtube.com')
    assert 'YouTube' in d.title
    assert "youtube" in d.current_url
    d.quit()

def test_wp():
    d = get_driver()
    d.get('https://www.wp.pl')
    assert "wp" in d.title.lower()
    assert d.find_element(By.CSS_SELECTOR, 'img')
    assert "wp" in d.current_url.lower()
    d.quit()

def test_booking():
    d = get_driver()
    d.get('https://www.booking.com')
    assert d.find_element(By.CSS_SELECTOR, 'img')
    assert d.find_element(By.NAME, 'ss')
    assert d.find_element(By.CSS_SELECTOR, 'button[type="submit"]')
    d.quit()

def test_interia():
    d = get_driver()
    d.get('https://www.interia.pl')
    assert d.find_element(By.TAG_NAME, 'footer')
    assert d.find_element(By.TAG_NAME, 'nav')
    assert d.find_element(By.TAG_NAME, 'header')
    d.quit()

def test_twitter():
    d = get_driver()
    d.get('https://twitter.com')
    assert 'X' in d.title or 'Twitter' in d.title
    assert d.find_element(By.TAG_NAME, 'svg')
    assert d.find_element(By.TAG_NAME, 'body').is_displayed()
    d.quit()

def test_stackoverflow():
    d = get_driver()
    d.get('https://stackoverflow.com')
    assert 'Stack Overflow' in d.title
    assert d.find_element(By.TAG_NAME, 'nav')
    assert d.find_element(By.TAG_NAME, 'header')
    d.quit()

def test_linkedin():
    d = get_driver()
    d.get('https://www.linkedin.com')
    assert d.find_element(By.CSS_SELECTOR, 'img')
    assert d.find_element(By.TAG_NAME, 'form')
    assert len(d.find_elements(By.TAG_NAME, 'a')) > 5
    d.quit()

def test_reddit():
    d = get_driver()
    d.get('https://www.reddit.com')
    assert d.find_element(By.TAG_NAME, 'header')
    assert d.find_elements(By.CSS_SELECTOR, '[data-testid="community-icon"]')
    assert d.find_elements(By.CSS_SELECTOR, 'div[data-testid="post-container"]')
    d.quit()

def test_medium():
    d = get_driver()
    d.get('https://medium.com')
    assert 'Medium' in d.title
    assert d.find_element(By.TAG_NAME, 'header')
    assert d.find_elements(By.TAG_NAME, 'section')
    d.quit()

def test_netflix():
    d = get_driver()
    d.get('https://www.netflix.com')
    assert d.find_element(By.CSS_SELECTOR, 'img')
    assert any('Zaloguj siÄ™' in a.text or 'Sign In' in a.text for a in d.find_elements(By.TAG_NAME, 'a'))
    assert len(d.find_elements(By.TAG_NAME, 'a')) > 3
    d.quit()

def test_imdb():
    d = get_driver()
    d.get('https://www.imdb.com')
    assert len(d.find_elements(By.TAG_NAME, 'a')) > 2
    assert "imdb" in d.title.lower()
    assert "imdb" in d.current_url.lower()
    d.quit()

def test_facebook():
    d = get_driver()
    d.get('https://www.facebook.com')
    assert 'Facebook' in d.title
    assert d.find_element(By.CSS_SELECTOR, 'img[alt="Facebook"]')
    assert d.find_element(By.NAME, 'email')
    d.quit()

def test_amazon():
    d = get_driver()
    d.get('https://www.amazon.com')
    assert 'Amazon' in d.title
    assert 'amazon' in d.page_source.lower()
    assert 'com' in d.page_source.lower()
    d.quit()

def test_olx():
    d = get_driver()
    d.get('https://www.olx.pl')
    assert d.find_element(By.CSS_SELECTOR, 'a')
    assert "olx" in d.current_url.lower()
    assert "olx" in d.title.lower()
    d.quit()
```

```python
test_cases = [
    test_onet, test_example, test_wikipedia, test_github,
    test_google, test_youtube, test_wp, test_booking, test_interia,
    test_twitter, test_stackoverflow, test_linkedin, test_reddit, test_medium,
    test_netflix, test_imdb, test_facebook, test_amazon, test_olx
]

results = {}
for test in test_cases:
    name = test.__name__
    try:
        test()
        results[name] = 'OK'
    except Exception as e:
        results[name] = f'FAIL: {e}'

import pprint
pprint.pprint(results)
```
